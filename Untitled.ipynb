### 线性回归

```python
# We will use a simple training set
X = 2 * np.random.rand(500, 1)
y = 5 + 3 * X + np.random.randn(500, 1)
fig = plt.figure(figsize=(8,6))
plt.scatter(X, y)
plt.title("Dataset")
plt.xlabel("First feature")
plt.ylabel("Second feature")
plt.show()
```

```python
# Split the data into a training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y)

print 'Shape X_train: ', X_train.shape
print 'Shape y_train: ', y_train.shape
print 'Shape X_test: ', X_test.shape
print 'Shape y_test: ', y_test.shape
```

```python
class MyLinearRegression(object):
    def __init__(self):
        pass
    
    def train_gradient_descent(self, X, y, learning_rate = 0.01, n_iters=100):
        n_samples, n_features = X.shape
        self.weights = np.zeros(shape=(n_features, 1))
        self.bias = 0
        costs = []
        
        for i in range(n_iters):
            # cal predict
            y_predict = np.dot(X, self.weights) + self.bias
            
            # cal loss
            cost = 0.5 / n_samples * np.sum((y_predict - y)**2)
            costs.append(cost)
            
            if i % 100 == 0:
                print 'cost at iteration {}: {}'.format(i, cost)
            
            # cal gradients
            dj_w = 1.0 / n_samples * np.dot(X.T, (y_predict - y))
            dj_b = 1.0 / n_samples * np.sum(y_predict - y)
            
            # update parameters
            self.weights -= learning_rate * dj_w
            self.bias -= learning_rate * dj_b
        
        return self.weights, self.bias, costs
    
    def predict(self, X):
        return np.dot(X, self.weights) + self.bias
```

### train linear

```python
n_iters=600
regressor = MyLinearRegression()
w_trained, b_trained, costs = regressor.train_gradient_descent(X_train, y_train, learning_rate=0.005, n_iters=n_iters)
fig = plt.figure(figsize=(8,6))
plt.plot(np.arange(n_iters), costs)
plt.title("Development of cost during training")
plt.xlabel("Number of iterations")
plt.ylabel("Cost")
plt.show()
```

### test data

```python
n_samples, _ = X_train.shape
n_samples_test, _ = X_test.shape

y_p_train = regressor.predict(X_train)
y_p_test = regressor.predict(X_test)

error_train =  (1.0 / n_samples) * np.sum((y_p_train - y_train) ** 2)
error_test =  (1.0 / n_samples_test) * np.sum((y_p_test - y_test) ** 2)

print "Error on training set: {}".format(np.round(error_train, 4))
print "Error on test set: {}".format(np.round(error_test))

fig = plt.figure(figsize=(8,6))
plt.scatter(X_train, y_train)
plt.scatter(X_test, y_p_test, c='r')
plt.xlabel("First feature")
plt.ylabel("Second feature")
plt.show()
```

```python
class MyLogisticRegression(object):
    def __init__(self):
        pass
    
    def sigmod(self, raw):
        return 1 / (1 + np.exp(-raw))
    
    def train(self, X, y_true, n_iters, learning_rate):
        n_samples, n_features = X.shape
        self.weights = np.zeros((n_features, 1))
        self.bias = 0
        costs = []
        
        for i in range(n_iters):
            y_pre = self.sigmod(np.dot(X, self.weights) + self.bias)
            cost = (-1.0 / n_samples) * np.sum(y_true * np.log(y_pre) + (1-y_true) * np.log(1-y_pre))
            
            dj_w = 1.0 / n_samples * learning_rate * np.dot(X.T, y_pre - y_true)
            dj_b = 1.0 / n_samples * learning_rate * np.sum(y_pre - y_true)
            
            self.weights -= dj_w
            self.bias -= dj_b
            
            costs.append(cost)
            
            if i % 100 == 0:
                print 'cost after iteration {} : {}'.format(i, cost)
#                 print dj_w
                
        return self.weights, self.bias, costs
        
    def predict(self, X):
        y_pre = self.sigmod(np.dot(X, self.weights) + self.bias)
        y_pre_label = [1 if v > 0.5 else 0 for v in y_pre]
        
        return np.array(y_pre_label)[:, np.newaxis]
```

```python
regressor = MyLogisticRegression()
w_trained, b_trained, costs = regressor.train(X_train, y_train, n_iters=600, learning_rate=0.009)
print len(costs)

fig = plt.figure(figsize=(8,6))
plt.plot(np.arange(600), costs)
plt.title("Development of cost over training")
plt.xlabel("Number of iterations")
plt.ylabel("Cost")
plt.show()
```

### k-means

```python
X, y = make_blobs(centers=4, n_samples=1000)
print('Shape of dataset: {}'.format(X.shape))

fig = plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=y)
plt.title("Dataset with 4 clusters")
plt.xlabel("First feature")
plt.ylabel("Second feature")
plt.show()
```

```python
class MyKMeans(object):
    def __init__(self, n_clusters=2):
        self.k = n_clusters
    
    def fit(self, X):
        n_samples, _ = X.shape
        self.centers = np.array(random.sample(list(X), self.k))
        self.initial_centers = np.copy(self.centers)
        
        old_assigns = None
        n_iters = 0
        
        while True:
            new_assigns = [self.classify(point) for point in X]
            
            if new_assigns == old_assigns:
                print n_iters,'iters finished!'
                return
            
            old_assigns = new_assigns
            n_iters += 1
            
            for id_ in range(self.k):
                idx = np.where(np.array(new_assigns) == id_)
                points = X[idx]
                self.centers[id_] = points.mean(axis=0)

    def l2_distance(self, point):
        dists = np.sqrt(np.sum((self.centers - point)**2, axis=1 ))
        return dists
    
    def classify(self, point):
        dists = self.l2_distance(point)
        return np.argmin(dists)
    
    def plot_clusters(self, data):
        plt.figure(figsize=(12,10))
        plt.title("Initial centers in black, final centers in red")
        plt.scatter(data[:, 0], data[:, 1], marker='.', c=y)
        plt.scatter(self.centers[:, 0], self.centers[:,1], c='r')
        plt.scatter(self.initial_centers[:, 0], self.initial_centers[:,1], c='y')
        plt.show()
```

```python
kmeans = MyKMeans(n_clusters=4)
kmeans.fit(X)
```

```python
kmeans.plot_clusters(X)
```

```python
class KMeans(object):
    def __init__(self, n_clusters=5):
        self.k = n_clusters
        
    def fit(self, x):
        self.centers = np.array(random.sample(list(x), self.k))
        
        old_res = None
        while True:
            new_res = [self.classify(v) for v in list(x)]
            
            if old_res == new_res:
                return 
            old_res = new_res
            
            for i in range(self.k):
                idxs = np.where(np.array(new_res) == i)[0]
#                 print idxs
                self.centers[i] = x[idxs].mean(axis=0)
    
    def classify(self, x):
        dists = np.sum((x - self.centers) ** 2, axis=1)
        return np.argmin(dists)
                
    def plot_clusters(self, data):
#         print self.centers
        plt.figure(figsize=(12,10))
        plt.title("Initial centers in black, final centers in red")
        plt.scatter(data[:, 0], data[:, 1], marker='.', c=y)
        plt.scatter(self.centers[:, 0], self.centers[:,1], c='r')
#         plt.scatter(self.initial_centers[:, 0], self.initial_centers[:,1], c='k')
        plt.show()
```

```python
def max_incre(raw):
    size = len(raw)
    dp = [1] * size
    for i in range(size):
        for j in range(i):
            if raw[i] > raw[j]:
                dp[i] = max(dp[i], dp[j]+1)
    return max(dp)
```
